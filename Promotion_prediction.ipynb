```python
data["total_score"]=data.no_of_trainings * data.avg_training_score
data.head()
# total_score is an entry based on how many trainings were attended and how well the person performed on average.
```

* conclusion from the plot is :
 Employees promoted have scores in the Medium and High range i.e. 65 and greater

 total score is not the only criterion for promotion.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler


# Original data
X = data[['KPIs_met', 'awards_won', 'avg_training_score', 'previous_year_rating']]
y = data['is_promoted']

ros=RandomOverSampler(random_state=1)
X_res,y_res=ros.fit_resample(X,y)
print(X_res.shape)
print(y_res.value_counts())
```

```python
# Train-test split on original data
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score,accuracy_score,classification_report,confusion_matrix


lr = LogisticRegression(class_weight="balanced",max_iter=1000)
lr.fit(X_train, y_train)
```

```python
# RandomClassifier..
rf=RandomForestClassifier(n_estimators=100,class_weight="balanced")
rf.fit(X_train, y_train)
```

```python
# XGB Classifier..
xgb=XGBClassifier(n_estimators=100,use_label_encoder=False,eval_metric='logloss')
xgb.fit(X_train, y_train)
```

```python
# Evaluation

for model in [lr,rf,xgb]:
    y_pred=model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    print(model.__class__.__name__,"\n")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("AUC-ROC:", roc_auc_score(y_test, y_prob))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n")
```

```python
y_prob = model.predict_proba(X_test)[:, 1]
y_prob
```

```python
models = [rf, xgb, lr]
model_names = ['Random Forest', 'XGBoost', 'Logistic Regression']

plt.figure(figsize=(15, 4))

for i, model in enumerate(models):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    
    plt.subplot(1, 3, i+1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_names[i]}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')

plt.suptitle('Confusion Matrices Comparison')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve,auc
plt.figure(figsize=(8, 6))

for model in [rf, xgb, lr]:
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{model.__class__.__name__} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid()
plt.tight_layout()

plt.show()
```
